```python
import numpy as np
import sounddevice as sd
from time import time
import soundfile as sf
from IPython.display import Audio 
from scipy.io.wavfile import write
```

```python
def micCallback(indata, frames, time, status):
    global data
    if status:
        print('Status:', status)
    data = np.hstack((data, np.transpose(indata)[0]))
```

```python
t0 = time()
sr = 44100
data = np.array([])
with sd.InputStream(samplerate=sr, channels=1, device=6, callback=micCallback, latency='high'): #, device=6
    while time() - t0 < 10:
        sd.sleep(50)
```

```python
print(data.size)
print(data.size / sr)
write('test.wav', sr, data)
#Audio(np.transpose(data),rate=sr)
```
