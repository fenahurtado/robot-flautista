```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import InterpolatedUnivariateSpline
```

```python
F = [0, 1000,1500,2000,2500,3000,3500,4000,4500,5000,5500,10000,22050]
M = [0.,2.85,2.49,1.65,1.55,1.81,1.35,1.00,1.13,1.58,1.21,3,0.]

gap_size = np.diff(F) // np.diff(F).min()
F_new = []
for i in range(len(F)-1):
    F_new.extend(np.linspace(F[i], F[i+1], gap_size[i], endpoint=False))
F_new.append(F[-1])

pl_spline = InterpolatedUnivariateSpline(F, M, k=1)
M_new = pl_spline(F_new)
smooth_spline = InterpolatedUnivariateSpline(F_new, M_new, k=5)

ff = np.linspace(F[0], F[-1], 100)
plt.plot(F, M, 'ro')
plt.plot(ff, smooth_spline(ff), 'b')
plt.show()
```
